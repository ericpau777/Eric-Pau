```python
import numpy as np
import random
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def process(hat):
    # Illustrates the process of randomly choosing two numbers from the hat 
    # and placing a paper with their difference into the hat.
    # Returns the hat after the process.
    num1 = random.choice(hat)
    hat.remove(num1)
    num2 = random.choice(hat)
    hat.remove(num2)
    hat.append(abs(num1 - num2))
    return hat
```

```python
process([1,2,3])
```

```python
def problem(n):
    # Encapsulate the whole problem, where n = 2024 in the problem
    hat = list(range(1,n+1))
    for _ in range(n - 1):
        process(hat)
    return hat
```

```python
problem(6)
```

```python
def Average(keys, probs):
    ttl = 0
    for i in range(len(keys)):
        ttl += keys[i] * probs[i]
    return ttl/len(keys)

def simulate(n, t, mean):
    # Runs the process for 1,...,n for t times
    res = []
    for _ in range(t):
        res.append(problem(n)[0])
    count = Counter(res)
    count_keys = list(count.keys())
    count_vals = list(count.values())
    count_probs = [x / t for x in count_vals]
    print("S = ", count_keys)
    print("P = ", count_probs)
    ttl = sum(res)
    ave = ttl / t
    print("Average = ", ave)
    mean.append(ave)

mean = []
for i in range(2, 17):
    print("n = ", i)
    simulate(i, 10000, mean)
    print("-----------------------------")
```
